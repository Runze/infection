```python
import networkx as nx
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 1. Function to convert inputs into a directed graph object

The function takes the following arguments:
1. **inputs** (dict): a dictionary with keys representing coaches and values students (e.g., {coach_1: [student_1, student_2], ...}). Both coaches and students are identified by a unique identifier, which can be either integers or strings.
2. **attr_name** (str; optional): the name of an attribute (currently only one attribute is implemented for each user). Defaulted to "version," indicating website version.
3. **default_attr_value** (int; optional): the default attribute value given to all users. Defaulted to 0, indicating Version 0 of the website.

```python
def create_network(inputs, attr_name = 'version', 
                   default_attr_value = 0):
    G = nx.DiGraph(inputs)

    # Set all users' attributes to the default value
    nodes = G.nodes()
    attrs = dict(zip(nodes, np.repeat(default_attr_value, len(nodes))))
    nx.set_node_attributes(G, attr_name, attrs)
    
    return G
```

### 2. Function to plot networks with infected nodes

The function plots the coaching graph and prints out all the connected components within the graph. It takes the following arguments:
1. **G** (networkx graph object): the network graph created by create_network.
2. **attr_name** (str; optional): the name of the attribute. Must be the same as the one given in create_network.
3. **attr_value_range** (list; optional): the range of all possible attribute values. Used to set color schemes in plotting (i.e., a higher attribtue value corresponds to a darker shade). If missing, [0, 1] is used.
4. **verbose** (bool; optional): whether to print out the description of the connected components.

The function doesn't return any value.

```python
def plot_network(G, attr_name = 'version', attr_value_range = [0, 1], verbose = True):
    nodes = G.nodes()
    nx.draw_circular(G, node_color = [nx.get_node_attributes(G, attr_name)[n] for n in nodes],
                     cmap = plt.cm.Blues, vmin = min(attr_value_range) - .5, vmax = max(attr_value_range) + .5)
    
    # Print out the weakly-connected components
    if verbose:
        comps = nx.weakly_connected_components(G)
        print 'There are %d weakly-connected components. They are:' % len(comps)
        for i, c in enumerate(comps):
            print '\tComponent', i, c
```

### 3. Total infection

Given any user, the function finds "the entire connected component of the coaching graph containing that user" and resets their attributes to a new value, indicating infection. It takes the following arguments:
1. **G** (networkx graph object): the network graph created using create_network.
2. **node** (int or str): the given user, identified using its unique identifier (must be the same as the one given in create_network).
3. **attr_name** (str; optional): the name of the attribute. Must be the same as the one given in create_network.
4. **infected_attr_value** (int; optional): the new attribute value to be given to the infected users. Defaulted to 1, indicating Version 1 of the website.
5. **return_new_graph** (bool; optional): whether to return a new graph object with the new attribtue value set for the infected or to return only a list of infected users.

```python
def total_infection(G, node, attr_name = 'version', infected_attr_value = 1, return_new_graph = True):
    # Identify weakly-connected components
    comps = nx.weakly_connected_components(G)
    
    # Find the sub-component with the node in it
    # Since a node can only appear in one connected component, a generator function is used that stops at the first match.
    infected = (c for c in comps if node in c).next()
    print 'Using total infection given the node %s, %d nodes are infected.' % (node, len(infected))
    print 'They are:', infected
    
    if return_new_graph:
        # Reset attributes for the infected node
        attrs_new = dict(zip(infected, np.repeat(infected_attr_value, len(infected))))
        G_new = G.copy()
        nx.set_node_attributes(G_new, attr_name, attrs_new)    
    
        return G_new
    else:
        return infected
```

### Limited infection

Given a number of users, the function tries to infect close to that number of users within the acceptable difference while keeping the connected components as intact as possible.


To do that, it starts with the smallest component (i.e., those with the fewest users) and infects more, one component at a time, in the ascending order until the goal is reached or about to be exceeded.


For example, if a network has 3 components with sizes being 3, 4, and 5. To reach a goal of 9, it will first infect 3, then 4, and stop at 5. Hence, it will miss the goal by 2 but, depending on the tolerance level, it may still declare a success.


One advantage of infecting in the ascending order is that more coaching groups can be infected this way, which will potentially make our treatment group more diverse and reduce selection bias.


After this process, if the number of infected users is still far from the goal and the allow_split option is turned on, additional splitting is performed *inside* the component that cannot be infected as a whole (i.e., infecting it all would have greatly exceeded the goal given the tolerance).


In this additional splitting, *immediate*, not transitive, coach-student relationships are kept intact. In other words, at this stage, the algorithm tries to infect the individual 'sub-trees' that make up the component, one at a time, also in the ascending order.


To continue the example above, now we are at the component with 5 nodes. Say it's made up of 2 sub-trees, one with size 4 and one with size 2 (e.g., Master Coach -> Student 1 + Student 2 + Sub Coach, Sub Coach -> Student 3), the 2-node sub-tree will be infected (i.e., Sub Coach -> Student 2). The rationale is that doing so will at least give some coaches and their direct students the same version of the website. If the structure of the network is such that the number of students per coach decreases as one moves down the hierarchy, this strategy will infect mostly bottom groups with fewer students receving shared coaching seeing different websites.


If doing so still cannot reach the goal, the algorithm gives up and returns whatever it has infected so far.


The function takes the following arguments:
1. **G** (networkx graph object): the network graph created using create_network.
2. **infection_goal** (int): the desired number of users to be infected.
3. **tol** (int; optional): the acceptable deviance from the infection goal; defaulted to 1.
4. **allow_split** (bool; optional): whether to allow in-component splitting (see description above).
5. **attr_name** (str; optional): the name of the attribute. Must be the same as the one given in create_network.
6. **infected_attr_value** (int; optional): the new attribute value to be given to the infected users. Defaulted to 1.
7. **return_new_graph** (optional): whether to return a new graph object with the new attribtue value set for the infected or to return only a list of infected users.

```python
def limited_infection(G, infection_goal, tol = 1, allow_split = True, attr_name = 'version', infected_attr_value = 1, return_new_graph = True):
    # Identify weakly-connected components in graphs
    # Note here, for purpose of further splitting, the 'comps' are graph objects instead of lists of nodes like in total_infection.
    comps = nx.weakly_connected_component_subgraphs(G)
    
    # Infect the components one at a time in the ascending order
    comps = sorted(comps, key = lambda x: len(x))
    infected = []
    
    for comp in comps:
        dist_to_goal = infection_goal - len(infected)
        if dist_to_goal != 0 and dist_to_goal - len(comp) >= -1:
            infected += comp.nodes()
        else:
            break
    
    print 'Using limited infection with a goal of %d (+/-%d), the infected nodes are:' % (infection_goal, tol)
    print '\tNodes selected from complete components:', infected
    
    # If the number of infected users is still far from the goal and further splitting is allowed, proceed as described above.
    if dist_to_goal > tol:
        if allow_split:
            # Make a copy of the current infected list
            infected_old = infected[:]

            # For each node in the component, pair it with its immediate successors.
            c_sub = zip(comp.nodes(), comp.adjacency_list())
            
            # Removing single child nodes (it's not good to infect only the student but not the coach)
            c_sub = [c for c in c_sub if len(c[1]) > 0]
            
            # sort by the size of the students in the ascending order
            c_sub = sorted(c_sub, key = lambda x: len(x[1]))

            # Right now each sub-tree is represented as (coach, [student_1, student_2, ...]).
            # For easy processing, convert them into lists as [coach, student_1, student_2, ...]
            # and remove duplicate nodes (e.g., one student can be coach in another group or one can be coached by 2 coaches).
            c_sub = [[c[0]] + c[1] for c in c_sub]
            c_sub = [list(set(c_sub[i]) - set(sum(c_sub[:i], []))) for i in xrange(len(c_sub))]

            # Similar as above, infect the sub-trees.
            for c in c_sub:
                dist_to_goal = infection_goal - len(infected)
                if dist_to_goal != 0 and dist_to_goal - len(c) >= -1:
                    infected += c
                else:
                    break

            print '\tNodes selected from sub-trees inside a complete component:', [node for node in infected if node not in infected_old]
    
        # If additional splitting is not allowed, give up and print out the failure message.
        else:
            print '\nWarning: Further splitting within a component is not allowed. Failed to reach the desired amount within the tolerance.' 
    
    # Summarize the results.
    print '\nThe total number of infected nodes is %d,' % len(infected)
    if dist_to_goal == 0:
        print 'same as the desired amount.'
    elif dist_to_goal < 0:
        print '%d more than the desired amount.' % abs(dist_to_goal)
    else:
        print '%d fewer than the desired amount.' % abs(dist_to_goal)
    
    if return_new_graph:
        # Reset attributes for the infected node
        attrs_new = dict(zip(infected, np.repeat(infected_attr_value, len(infected))))
        G_new = G.copy()
        nx.set_node_attributes(G_new, attr_name, attrs_new)
        
        return G_new
    else:
        return infected
```

### Tests!

```python
# Test case: a network with multiple weakly-connected components
inputs_1 = {0: [1, 2, 3], 4: [3, 5], 5: [6], 7: [8, 9], 10: [11, 12, 13], 13: [14, 15]}

# Create network graph
G1 = create_network(inputs_1)

# Plot the original network
print 'The orignial, uninfected, network:'
plot_network(G1)
```

```python
# Total infection given node 13
G1_total = total_infection(G1, 13)
plot_network(G1_total, verbose = False)
```

```python
# Limited infection given a goal of 10
G1_limited = limited_infection(G1, 10)
plot_network(G1_limited, verbose = False)
```

```python
# Limited infection given a goal of 11
G1_limited = limited_infection(G1, 11)
plot_network(G1_limited, verbose = False)
```

```python
# Limited infection given the goal of 11 but disallowing in-component splitting
# Note this is my understanding of "the exact version of limited infection."
G1_limited_no_split = limited_infection(G1, 11, allow_split = False, return_new_graph = False)
```
